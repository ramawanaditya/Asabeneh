```python
#soal nomer 4

num_one = 5
num_two = 4

total =(num_one + num_two)
print("Hasilnya penjumlahannya adalah =", total)
diff = (num_one - num_two)
print("Hasil penguranganya adalah =", diff)
produk = (num_one * num_two)
print("Hasil perkaliannya adalah =", produk)
divisi = (num_one / num_two)
print("Hasil pembagiannya adalah =", divisi)
sisa = (num_one % num_two)
print("Hasil persenan adalah =", sisa)
exp = (num_one ** num_two)
print("Hasil pangkat adalah =", exp)
floor_division = (num_one // num_two)
print("Hasil dari Floor division adalah =", floor_division)
```
